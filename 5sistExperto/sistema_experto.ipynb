```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Todos los antecendentes son relativos a las plantas
color_hojas = ctrl.Antecedent(np.arange(0, 101, 1), 'color_hojas')
num_insectos = ctrl.Antecedent(np.arange(0, 101, 1), 'num_insectos')
humedad_suelo = ctrl.Antecedent(np.arange(0, 101, 1), 'humedad_suelo')
calidad_raices = ctrl.Antecedent(np.arange(0, 101, 1), 'calidad_raices')
velocidad_crecimiento = ctrl.Antecedent(np.arange(0, 101, 1), 'velocidad_crecimiento')
num_manchas = ctrl.Antecedent(np.arange(0, 101, 1), 'num_manchas')
num_deformaciones = ctrl.Antecedent(np.arange(0, 101, 1), 'num_deformaciones')
tamaño_tallo = ctrl.Antecedent(np.arange(0, 101, 1), 'tamaño_tallo')
intensidad_sol = ctrl.Antecedent(np.arange(0, 101, 1), 'intensidad_sol')


# Todos los consecuentes son relativos a las plantas
deficiencia_nitrogeno = ctrl.Consequent(np.arange(0, 101, 1), 'deficiencia_nitrogeno')
deficiencia_potasio = ctrl.Consequent(np.arange(0, 101, 1), 'deficiencia_potasio')
deficiencia_fosforo = ctrl.Consequent(np.arange(0, 101, 1), 'deficiencia_fosforo')
infestacion_plagas = ctrl.Consequent(np.arange(0, 101, 1), 'infestacion_plagas')
infeccion_hongos = ctrl.Consequent(np.arange(0, 101, 1), 'infeccion_hongos')
exceso_agua = ctrl.Consequent(np.arange(0, 101, 1), 'exceso_agua')
falta_agua = ctrl.Consequent(np.arange(0, 101, 1), 'falta_agua')
quemadura_solar = ctrl.Consequent(np.arange(0, 101, 1), 'quemadura_solar')

color_hojas.automf(3)
num_insectos.automf(3)
humedad_suelo.automf(3)
calidad_raices.automf(3)
velocidad_crecimiento.automf(3)
num_manchas.automf(3)
num_deformaciones.automf(3)
tamaño_tallo.automf(3)
intensidad_sol.automf(3)

deficiencia_nitrogeno.automf(3)
deficiencia_potasio.automf(3)
deficiencia_fosforo.automf(3)
infestacion_plagas.automf(3)
infeccion_hongos.automf(3)
exceso_agua.automf(3)
falta_agua.automf(3)
quemadura_solar.automf(3)
```

Reglas

```python
regla_def_nitrogeno1 = ctrl.Rule(
    color_hojas['poor'] & velocidad_crecimiento['poor'],
    deficiencia_nitrogeno['good']
)

regla_def_nitrogeno2 = ctrl.Rule(
    color_hojas['poor'],
    deficiencia_nitrogeno['average']
)

regla_def_nitrogeno3 = ctrl.Rule(
    color_hojas['average'],
    deficiencia_nitrogeno['poor']
)

regla_def_potasio1 = ctrl.Rule(
    (num_manchas['good'] & num_deformaciones['average']) | num_manchas['good'] & num_deformaciones['good'],
    deficiencia_potasio['good']
)

regla_def_potasio2 = ctrl.Rule(
    num_manchas['average'] & num_deformaciones['average'],
    deficiencia_potasio['average']
)

regla_def_potasio3 = ctrl.Rule(
    num_manchas['average'],
    deficiencia_potasio['poor']
)

regla_def_fosforo1 = ctrl.Rule(
    calidad_raices['poor'] & tamaño_tallo['poor'],
    deficiencia_fosforo['good']
)

regla_def_fosforo2 = ctrl.Rule(
    (calidad_raices['poor'] & tamaño_tallo['average']) | calidad_raices['average'] & tamaño_tallo['poor'],
    deficiencia_fosforo['average']
)

regla_def_fosforo3 = ctrl.Rule(
    calidad_raices['average'],
    deficiencia_fosforo['poor']
)

regla_plagas1 = ctrl.Rule(
    (num_insectos['good'] & num_deformaciones['good']) | (num_insectos['good'] & num_deformaciones['average']) | num_insectos['average'] & num_deformaciones['good'],
    infestacion_plagas['good']
)

regla_plagas2 = ctrl.Rule(
    (num_insectos['average'] & num_deformaciones['average']),
    infestacion_plagas['average']
)

regla_plagas3 = ctrl.Rule(
    num_insectos['average'],
    infestacion_plagas['poor']
)

regla_hongos1 = ctrl.Rule(
    (num_manchas['poor'] & humedad_suelo['poor']) | num_manchas['average'] & humedad_suelo['poor'],
    infeccion_hongos['good']
)

regla_hongos2 = ctrl.Rule(
    (num_manchas['average'] & humedad_suelo['average']),
    infeccion_hongos['average']
)

regla_hongos3 = ctrl.Rule(
    num_manchas['average'],
    infeccion_hongos['poor']
)

regla_falta_agua1 = ctrl.Rule(
    color_hojas['poor'] & humedad_suelo['poor'],
    falta_agua['good']
)

regla_falta_agua2 = ctrl.Rule(
    color_hojas['poor'] & humedad_suelo['average'],
    falta_agua['average']
)

regla_falta_agua3 = ctrl.Rule(
    humedad_suelo['average'],
    falta_agua['poor']
)

regla_exceso_agua1 = ctrl.Rule(
    color_hojas['poor'] & humedad_suelo['good'],
    exceso_agua['good']
)

regla_exceso_agua2 = ctrl.Rule(
    color_hojas['average'] & humedad_suelo['good'],
    exceso_agua['average']
)

regla_exceso_agua3 = ctrl.Rule(
    color_hojas['average'],
    exceso_agua['poor']
)

regla_quemadura1 = ctrl.Rule(
    (num_manchas['good'] | num_manchas['average']) & intensidad_sol['good'],
    quemadura_solar['good']
)

regla_quemadura2 = ctrl.Rule(
    num_manchas['good'] & intensidad_sol['average'],
    quemadura_solar['average']
)

regla_quemadura3 = ctrl.Rule(
    intensidad_sol['average'],
    quemadura_solar['poor']
)

```

sistemas de control

```python
control_nitrogeno = ctrl.ControlSystem([regla_def_nitrogeno1, regla_def_nitrogeno2, regla_def_nitrogeno3])
control_potasio = ctrl.ControlSystem([regla_def_potasio1, regla_def_potasio2, regla_def_potasio3])
control_fosforo = ctrl.ControlSystem([regla_def_fosforo1, regla_def_fosforo2, regla_def_fosforo3])
control_plagas = ctrl.ControlSystem([regla_plagas1, regla_plagas2, regla_plagas3])
control_hongos = ctrl.ControlSystem([regla_hongos1, regla_hongos2, regla_hongos3])
control_falta_agua = ctrl.ControlSystem([regla_falta_agua1, regla_falta_agua2, regla_falta_agua3])
control_exceso_agua = ctrl.ControlSystem([regla_exceso_agua1, regla_exceso_agua2, regla_exceso_agua3])
control_quemaduras = ctrl.ControlSystem([regla_quemadura1, regla_quemadura2, regla_quemadura3])
```

Simulacion

```python
prob_def_nitrogeno = ctrl.ControlSystemSimulation(control_nitrogeno)
prob_def_potasio = ctrl.ControlSystemSimulation(control_potasio)
prob_def_fosforo = ctrl.ControlSystemSimulation(control_fosforo)
prob_plaga = ctrl.ControlSystemSimulation(control_plagas)
prob_hongos = ctrl.ControlSystemSimulation(control_hongos)
prob_falta_agua = ctrl.ControlSystemSimulation(control_falta_agua)
prob_exceso_agua = ctrl.ControlSystemSimulation(control_exceso_agua)
prob_quemaduras = ctrl.ControlSystemSimulation(control_quemaduras)
```

```python
prob_def_nitrogeno.input['color_hojas'] = 60
prob_def_nitrogeno.input['velocidad_crecimiento'] = 30

prob_def_nitrogeno.compute()

print("Probabilidad de deficiencia de nitrógeno:", prob_def_nitrogeno.output['deficiencia_nitrogeno'])
```
